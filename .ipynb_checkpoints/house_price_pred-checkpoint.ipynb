### One way ANOVA

a. H0 (null hypothesis): 1 = 2 = 3 =  = k (all the population means are equal)

b. H1 (alternate hypothesis): at least one population mean is different from the rest

```python
final_df['FireplaceQu'].value_counts()
```

```python
grouped_anova = train_df.groupby(['FireplaceQu'])
```

```python
### Since p-value is less than 0.05 we can reject the null hypothesis

import scipy as sc

anova_1 = sc.stats.f_oneway(grouped_anova.get_group('Gd')['SalePrice'] ,grouped_anova.get_group('TA')['SalePrice'] )
anova_1
```

```python
### Since there is not much statistical difference among the groups => fail to reject the null hypothesis

val = final_df['FireplaceQu'].value_counts().index

for i in range(4):
    for j in range(i+1,5):
        anova = sc.stats.f_oneway(grouped_anova.get_group(val[i])['SalePrice'] ,grouped_anova.get_group(val[j])['SalePrice'] )
        print("{} and {} : {}".format(val[i] , val[j] , anova))
```

```python
### returns those features that that highly correlated among each other based on the theshold

def correlation(dataset , threshold):
    corr_matrix = dataset.corr()
    cols= set()
    
    for i in range(len(corr_matrix)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                cols.add(colname)
    
    return cols
```

## Hyperparameter Tuning with Xgboost Regressor

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost
xgb = xgboost.XGBRegressor()
```

```python
def hyperParameterTuning(X_train, y_train):
    param_tuning = {
        'learning_rate': [0.01 , 0.05 , 0.1 , 0.15 , 0.2],
        'max_depth': [2, 5, 7, 10 , 15 , 20],
        'min_child_weight': [1, 2, 3,4 , 5 , 6],
        'subsample': [0.5, 0.7],
        'colsample_bytree': [0.5, 0.7],
        'booster': ['gbtree', 'gblinear'],
        'n_estimators' : [ 500 , 900 , 1100 , 1500 , 2000 , 2500 , 3000],
        'objective': ['reg:squarederror']
    }

    xgb_model = xgboost.XGBRegressor()

    rsearch = RandomizedSearchCV(estimator = xgb_model,
                           param_distributions = param_tuning,                        
                           #scoring = 'neg_mean_absolute_error', #MAE
                           scoring = 'neg_root_mean_squared_error',  #RMSE
                           cv = 5,
                           n_iter = 100,
                           n_jobs = -1,
                           verbose = 2,
                           return_train_score = True,
                           random_state = 42)

    rsearch.fit(X_train,y_train)

    return rsearch.best_params_
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
X_train_new , X_test_new , y_train_new , y_test_new  = train_test_split(X_train , y_train_transform , random_state =0 , test_size = 0.2)
```

```python
dtrain = xgb.DMatrix(X_train_new, label=y_train_new)
dtest = xgb.DMatrix(X_test_new, label=y_test_new)
```

```python
# "Learn" the mean from the training data
mean_train = np.mean(y_train_new)

# Get predictions on the test set
baseline_predictions = np.ones(y_test_new.shape) * mean_train

# Compute MAE
rmse_baseline = np.sqrt(mean_squared_error(y_test_new, baseline_predictions))
print("Baseline RMSE is {:.2f}".format(rmse_baseline))
```

```python
params = {
    # Parameters that we are going to tune.
    'max_depth':6,
    'min_child_weight': 1,
    'eta':.3,
    'subsample': 1,
    'colsample_bytree': 1,
    # Other parameters
    'objective':'reg:squarederror',
}
params['eval_metric'] = "rmse"
num_boost_round = 999
```

```python
model = xgb.train(
    params,
    dtrain,
    num_boost_round=num_boost_round,
    evals=[(dtest, "Test")],
    early_stopping_rounds=20
)
```

```python
print("Best RMSE: {:.2f} with {} rounds".format(
                 model.best_score,
                 model.best_iteration+1))
```

```python
cv_results = xgb.cv(
    params,
    dtrain,
    num_boost_round=num_boost_round,
    seed=42,
    nfold=5,
    metrics={'rmse'},
    early_stopping_rounds=10
)
cv_results
```

```python
cv_results['test-rmse-mean'].min()
```

```python
# You can try wider intervals with a larger step between
# each value and then narrow it down. Here after several
# iteration I found that the optimal value was in the
# following ranges.
gridsearch_params = [
    (max_depth, min_child_weight)
    for max_depth in range(9,12)
    for min_child_weight in range(5,8)
]
```

```python
# Define initial best params and MAE
min_rmse = float("Inf")
best_params = None
for max_depth, min_child_weight in gridsearch_params:
    print("CV with max_depth={}, min_child_weight={}".format(
                             max_depth,
                             min_child_weight))
    # Update our parameters
    params['max_depth'] = max_depth
    params['min_child_weight'] = min_child_weight
    # Run CV
    cv_results = xgb.cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'rmse'},
        early_stopping_rounds=10
    )
    # Update best MAE
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin()
    print("\tRMSE {} for {} rounds".format(mean_rmse, boost_rounds))
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (max_depth,min_child_weight)
print("Best params: {}, {}, RMSE: {}".format(best_params[0], best_params[1], min_rmse))
```

```python
params['max_depth'] = 9
params['min_child_weight'] = 7
```

```python
gridsearch_params = [
    (subsample, colsample)
    for subsample in [i/10. for i in range(7,11)]
    for colsample in [i/10. for i in range(7,11)]
]
```

```python
min_rmse = float("Inf")
best_params = None
# We start by the largest values and go down to the smallest
for subsample, colsample in reversed(gridsearch_params):
    print("CV with subsample={}, colsample={}".format(
                             subsample,
                             colsample))
    # We update our parameters
    params['subsample'] = subsample
    params['colsample_bytree'] = colsample
    # Run CV
    cv_results = xgb.cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'rmse'},
        early_stopping_rounds=10
    )
    # Update best score
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin()
    print("\tRMSE {} for {} rounds".format(mean_rmse, boost_rounds))
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (subsample,colsample)
print("Best params: {}, {}, RMSE: {}".format(best_params[0], best_params[1], min_rmse))
```

```python
params['subsample'] = 1.0
params['colsample_bytree'] = 0.9
```

```python
%time
# This can take some time
min_rmse = float("Inf")
best_params = None
for eta in [.3, .2, .1, .05, .01, .005]:
    print("CV with eta={}".format(eta))
    # We update our parameters
    params['eta'] = eta
    # Run and time CV
    %time 
    cv_results = xgb.cv(
            params,
            dtrain,
            num_boost_round=num_boost_round,
            seed=42,
            nfold=5,
            metrics=['rmse'],
            early_stopping_rounds=10
    )
    # Update best score
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin()
    print("\tRMSE {} for {} rounds\n".format(mean_rmse, boost_rounds))
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = eta
print("Best params: {}, RMSE: {}".format(best_params, min_rmse))
```
